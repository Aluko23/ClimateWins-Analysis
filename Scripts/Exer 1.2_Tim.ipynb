# 2. Scaling

```python
#columns to exclude in the scaling
columns_to_exclude = ['DATE', 'MONTH']
```

```python
# Selecting columns to scale by excluding specified columns
features_to_scale = df.columns[~df.columns.isin(columns_to_exclude)]
```

```python
#create scaler object using StandardScaler from sklearn.preprocessing
#StandardScaler assumes data is normally distributed and scales with a distribution
#around 0 and standard deviation of 1. Scaling happens independently with each variable.

scaler = StandardScaler()
```

```python
# Fit and transform the features to scale
scaled_features = scaler.fit_transform(df[features_to_scale])
```

```python
# Create a new DataFrame with the scaled features
scaled_df = pd.DataFrame(scaled_features, columns=features_to_scale)
```

```python
#Checking the head for the scaled dataframe 'scaled_df'
scaled_df.head()
```

```python
#adding back the excluded columns
result_df = pd.concat([df[columns_to_exclude].reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)
```

```python
#Checking the head of the re-constituted dataframe after scaling
result_df.head()
```

```python
#export the scaled data in pickle form as [DATASET]_scaled
result_df.to_pickle(os.path.join(path, 'Processed Data', '[DATASET]_scaled.pkl'))
```

```python
#export the scaled data in pickle form as [DATASET]_scaled
result_df.to_csv(os.path.join(path, 'Processed Data', '[DATASET]_scaled.csv'))
```
