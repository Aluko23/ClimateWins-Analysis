# 3. ANN - Scripts 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier

#Scaling the data
scaler = StandardScaler()

#Fit the train data
scaler.fit(X_train)

#Scaling the X data
X_train_sc = scaler.transform(X_train)
X_test_sc = scaler.transform(X_test)

print(X_train_sc[:10])
```

```python
#Show scaling differences ##turn off legend
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))

ax1.set_title('Before Scaling')
sns.kdeplot(X_train, ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(X_train_sc, ax=ax2)

plt.show()
```

In these two neural network configurations:

Scenario 1 uses a simpler architecture with two hidden layers of 5 nodes each, 500 maximum iterations, and a tolerance of 0.0001. This setup aims for faster training with a reduced risk of overfitting, suitable for simpler datasets.
  
Scenario 2 has a more complex model with three hidden layers (10, 5, and 8 nodes), allowing up to 800 iterations and a stricter tolerance of 0.00001. This configuration may capture more detailed patterns but requires more training time and has a higher risk of overfitting.

In summary: Scenario 1 prioritizes efficiency and is better for simpler data, while Scenario 2 aims for depth and accuracy, fitting for more complex datasets.

```python
#SCENARIO 3 CODE
mlp = MLPClassifier(hidden_layer_sizes=(38, 18, 28), max_iter=2800, tol=0.00000000001)
#Fit the data to the model
mlp.fit(X_train, y_train)
```

The code tests the accuracy of two different Artificial Neural Network (ANN) models on training and test datasets, comparing how well the models perform on both sets of data:

1. ANN Model 1
   - The model is evaluated on both training data (`X_train`, `y_train`) and test data (`X_test`, `y_test`).
   - It achieves a similar accuracy score of *0.49%* on both datasets, indicating that the model's predictions are nearly random and not effective. This low accuracy suggests either inadequate model complexity or an issue with the dataset itself (e.g., non-linearity or noise).

2. ANN Model 2:
   - The code will also evaluate a second model (although results are not yet shown).
   - ANN Model 2’s architecture could potentially offer better accuracy, depending on the model's suitability for the data.

Given Model 1’s poor performance, adjustments such as increasing the number of nodes, adding more layers, or using a different preprocessing approach may improve the results. Additionally, examining the data quality and feature relevance could help identify potential issues with model accuracy.

Model Accuracy: Both ANN Model 1 and Model 2 show very low accuracy scores around 0.5% for training and test data.

Issues Observed: The extremely low accuracy suggests underfitting, meaning the model fails to capture patterns in the data.

```python
#testing ANN accuracy #ANN MODEL 3
y_pred = mlp.predict(X_train)
print(accuracy_score(y_pred, y_train))
y_pred_test = mlp.predict(X_test)
print(accuracy_score(y_pred_test, y_test))
```
